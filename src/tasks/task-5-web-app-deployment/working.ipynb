```python
def read_csv_folder(folder_path):
    """Reads all CSV files in a folder and creates separate dataframes for each file"""
    csv_files = sorted([f for f in os.listdir(folder_path) if f.endswith('.csv')])
    for i, file in enumerate(csv_files):
        globals()[str(i)] = pd.read_csv(os.path.join(folder_path, file))
```

```python
def get_data():
    data_folder = "../task-5-web-app-deployment/data/model"
    files = [f for f in os.listdir(data_folder) if f.endswith(".csv")]
    dfs = {}
    for file in files:
        df_name = os.path.splitext(file)[0]
        df = pd.read_csv(os.path.join(data_folder, file))
        df = df.iloc[:, :-2]  # remove last column
        dfs[df_name] = df.set_index(df.columns[0])
    return dfs
```
